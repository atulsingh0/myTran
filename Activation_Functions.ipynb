```python
import matplotlib.pyplot as plt
import numpy as np
import math

%matplotlib inline
```

### Activation Function List &nbsp;&nbsp;&nbsp;&nbsp; [Activation Function Table](https://en.wikipedia.org/wiki/Activation_function)

### Sigmoid Function  
## $$ \sigma(x)=f(x) = \frac{1}{1+e^{-x}} $$  

### Softmax Function  
## $$ \sigma(x_i)  = \frac{e^{x_i}}{\Sigma_{k=1}^n{e^{x_k}}}  $$   

### ReLU Function  
## $$ f(x)  = max(x,0)  = {\begin{cases}0&{\text {for }}x<0\\x&{\text{for }}x\geq0\end{cases}} $$  

### Binary Step Function  
## $$ {\displaystyle f(x)={\begin{cases}0&{\text{for }}x<0\\1&{\text{for }}x\geq 0\end{cases}}} $$  

### Tanh Function  
## $$ {\displaystyle f(x)=\tanh(x)={\frac {2}{1+e^{-2x}}}-1}  = 2\sigma(2x)-1 $$

```python
x= np.linspace(-10, 10, 100, endpoint=True)
y_sig = 1/(1+np.exp(-x))   # sigmoid
y_soft = np.exp(x)/np.sum(np.exp(x))  # softmax
y_relu = [np.maximum(x[i], 0) for i in np.arange(len(x))]  # ReLU function
y_bstep = [1 if i>=0 else 0 for i in x]
y_tanh = 2/(1+np.exp(-2*x)) - 1
```

```python
#fig=plt.figure(figsize=(10,8))

plt.plot(x,y_sig, linewidth=2, label='Sigmoid')
plt.plot(x,y_soft,linewidth=2, label='Softmax')
plt.plot(x,y_bstep, linewidth=2, label='Binary Step')

plt.ylim([-0.05, 1.3])
plt.legend(loc=0)
plt.grid(True)
```

```python
plt.plot(x,y_sig, linewidth=2, label='Sigmoid')
plt.plot(x,y_tanh, linewidth=2, label='TanH')

plt.ylim([-1.3, 1.3])
plt.legend(loc=0)
plt.grid(True)
```

```python
plt.plot(x,y_relu,linewidth=2, label='ReLU')

plt.ylim([-1.3, 11])
plt.legend(loc=0)
plt.grid(True)
```

```python
print("Mean of Sigmoiad Func:",np.sum(y_sig))
print("Mean of Softmax Func:",np.sum(y_soft))
print("Mean of Step Func:",np.sum(y_soft))
print("Mean of TanH Func:",np.sum(y_tanh))
print("Mean of ReLU Func:",np.sum(y_relu))
```

### Tips to choose activation function  
* For binary classification problem, Use Sigmoid activation (0,1) for output layer and tanH/ReLU function (-1,1) for hidden layer  
* Downsize of sigmoid and tanH function is - if input x or z is very large or small, the slope of activation function is very small and algorithm converge very slowly.  
* In this case, Use ReLU activation function but problem with ReLU function is, the derivatives/slope of function is 0 for -ve value. To overcome this issue use ReCluLU

```python
plt.plot(x,np.gradient(y_tanh), linewidth=2, label='TanH')
plt.plot(x,np.gradient(y_sig), linewidth=2, label='Sigmoid')

#plt.ylim([-0.3, 1.3])
plt.legend(loc=0)
plt.grid(True)
```
